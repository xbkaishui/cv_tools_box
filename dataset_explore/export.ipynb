```python
def filter_cat_by_ids(catIds, keep):
    imgIds = coco.getImgIds(catIds=catIds)[:keep]
    ann_ids = coco.getAnnIds(imgIds=imgIds, catIds=catIds)
    anns = coco.loadAnns(ids=ann_ids)
    images = coco.loadImgs(imgIds)
    return images, anns
```

```python
# coco.getCatIds()
import random

imgs = []
anns = []

img_id_set = set()
per_ann_set = set()

categories = [[1], [18], [17], [1,17], [1,18], [17,18]]
for cat in categories:
    if len(cat) == 1:
        keep = 150
    else:
        keep = 50
    per_imgs, per_anns = filter_cat_by_ids(cat, keep)
    for img in per_imgs:
        img_id = img['id']
        if img_id not in img_id_set:
            imgs.append(img)
            img_id_set.add(img_id)
    for ann in per_anns:
        ann_id = ann['id']
        if ann_id not in per_ann_set:
            anns.append(ann)
            per_ann_set.add(ann_id)    

output_path = './outputs'

import os
import shutil
shutil.rmtree(output_path, ignore_errors=True)
os.makedirs(output_path, exist_ok=True)


for img in imgs:
    img_path = os.path.join(output_path, img['file_name'])
    shutil.copyfile(os.path.join(coco_root, img['file_name']), img_path)
```
