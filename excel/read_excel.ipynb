```python
import openpyxl

# Specify the path to your Excel file
excel_file_path = '/Users/xbkaishui/Downloads/handian/Brunt/粘焊10→12.xlsx'

# Load the workbook
workbook = openpyxl.load_workbook(excel_file_path)

# Select the desired sheet
sheet = workbook.active  # or use workbook['SheetName'] to select a specific sheet by name

# Accessing data from cells
cell_value = sheet['A1'].value
print(f'Value in cell A1: {cell_value}')

max_column = sheet.max_column
max_column = 1 
# Iterating through rows and columns
for row in sheet.iter_rows(min_row=1, max_row=sheet.max_row, min_col=1, max_col=max_column):
    for cell in row:
        print(cell.value, end='\t')
    print()

# Close the workbook when done
workbook.close()
```
